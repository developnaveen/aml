```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,classification_report
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3 , random_state = 42)
```

```python
model = DecisionTreeClassifier(criterion = 'entropy', splitter = 'random',max_depth = 10)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_predict)
report = classification_report(y_test ,y_predict)
```

```python
print(f"\t \t \t Decision Tree Measures \n\n Accuracy : {accuracy} \n Classification Report : {report}")
```

```python
target_list = data.target_names.tolist()
```

```python
plt.figure(figsize = (12,15))
plot_tree(model,filled = True ,feature_names = data.feature_names , class_names = target_list , rounded = True)
plt.title("Decision Tree Algorithm Result")
plt.show()
```
