```python
import statistics
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
data = load_iris()
sepal_length = data.data[:,0]
print(sepal_length)
```

```python
mean = statistics.mean(sepal_length)
median = statistics.median(sepal_length)
mode = statistics.multimode(sepal_length)
variance = statistics.variance(sepal_length)
std = statistics.stdev(sepal_length)
```

```python
print(f"Statictical Analysis \n Mean : {mean} \n Mode : {mode} \n Median : {median} \n Variance : {variance} \n Standard deviation : {std}")
```

```python
plt.hist(sepal_length ,bins = 20 , color = "grey" ,edgecolor = "white" )
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequence")
plt.show()
```
