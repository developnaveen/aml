```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
data = load_diabetes()
print(f'Features : {data.feature_names} \n')
```

```python
X = data.data
y = data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=40)
```

```python
X_train_bmi = X_train[:,2].reshape(-1,1)
X_test_bmi = X_test[:,2].reshape(-1,1)
```

```python
lModel = LinearRegression().fit(X_train_bmi,y_train)
```

```python
l_predict = lModel.predict(X_test_bmi)
```

```python
mModel = LinearRegression().fit(X_train,y_train)
```

```python
m_predict = mModel.predict(X_test)
```

```python
lmae = mean_absolute_error(y_test,l_predict)
lrmse = np.sqrt(mean_squared_error(y_test,l_predict))
mmae = mean_absolute_error(y_test,m_predict)
mrmse = np.sqrt(mean_squared_error(y_test,m_predict))
```

```python
print(f'Linear Regression Model Accuracy \n MAE : {lmae} \n RMSE : {lrmse}')
```

```python
print(f'Multiple Linear Regression Model Accuracy \n MAE : {mmae} \n RMSE : {mrmse}')
```

```python
plt.figure(figsize = (10,5))
plt.scatter(y_test,l_predict,c='b',label='Linear Regression')
plt.scatter(y_test,m_predict,c='r',label='Multiple Regression')
plt.plot([0,350],[0,350],'--g',label='Perfect Prediction')
plt.title('Linear vs Multiple Linear Regression')
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.legend()
plt.show()
```
