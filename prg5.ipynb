```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
data = load_iris()
X = data.data
y = data.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 42)
```

```python
knn = KNeighborsClassifier().fit(X_train,y_train)
```

```python
y_predict_Knn = knn.predict(X_test)
```

```python
knn_accuracy = accuracy_score(y_test,y_predict_Knn)
```

```python
nb = GaussianNB().fit(X_train,y_train)
```

```python
y_predict_nb = nb.predict(X_test)
```

```python
nb_accuracy = accuracy_score(y_test,y_predict_nb)
```

```python
print(f"\t Accuracy \n KNN : {knn_accuracy} \n NB : {nb_accuracy}")
```

```python
plt.figure(figsize = (15,5))
plt.subplot(1,3,1)
plt.scatter(X_test[:,0],X_test[:,1], c = y_predict_Knn , cmap = 'viridis')
plt.title('KNN Prediced Lables')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.subplot(1,3,2)
plt.scatter(X_test[:,0],X_test[:,1],c = y_predict_nb , cmap = 'viridis')
plt.title('NB Predicted Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
